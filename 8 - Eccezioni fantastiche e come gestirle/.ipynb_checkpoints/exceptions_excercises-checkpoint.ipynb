## 1. Vettori Reloaded
Durante la nostra esercitazione con le classi abbiamo realizzato una classe per rappresentare vettori, in pi√π metodi abbiamo verificato che la lunghezza di due vettori fosse uguale. Rimpiazziamo l'if utilizzato in questi casi con un assert.

```python
class Vector():

    def __init__(self, values):
        self.values = values

        
    def __getitem__(self, i):
      return self.values[i]


    def __repr__(self) :
       return str(self.values)


    def __add__(self, vector):
        
      assert(len(self.values)==len(vector.values)), "I vettori hanno dimensioni diverse"
    
      list_sum = []

      for val1, val2 in zip(self.values, vector.values):
        list_sum.append(val1+val2)

      return Vector(list_sum)


    def __sub__(self, vector):

      assert(len(self.values)==len(vector.values)), "I vettori hanno dimensioni diverse"

      list_sum = []

      for val1, val2 in zip(self.values, vector.values):
        list_sum.append(val1-val2)

      return Vector(list_sum)


    def __eq__(self, vector):

      assert(len(self.values)==len(vector.values)), "I vettori hanno dimensioni diverse"

      for val1, val2 in zip(self.values, vector.values):
        if(val1 != val2):
          return False
      return True


    def __mul__(self, vector):
      return self.dot(vector)


    def sum(self):
      
      v_sum = 0

      for v in self.values:
        v_sum+=v

      return v_sum


    def dot(self, vector):

      assert(len(self.values)==len(vector.values)), "I vettori hanno dimensioni diverse"

      prod = 0

      for val1, val2 in zip(self.values, vector.values):
        prod+=val1*val2
      
      return prod


    def norm(self):
      return self.dot(self)**(1/2)



v1 = Vector([1, 2, 3])
print(v1)
print(v1[0])
print(v1.sum())

v2 = Vector([2, 4, 6])
print(v1 + v2)
print(v2 - v1)

print(v1==v1)
print(v1==v2)

print(v1*v2)

v3 = Vector([10, 20, 30, 40])
print(v1*v3)

print(v3.norm())
```

## 2. Calcolatrice

```python
def calculator(a, b, op):
    if(op=="+"):
        return a+b
    elif(op=="-"):
        return a-b
    elif(op=="*"):
        return a*b
    elif(op=="/"):
         try:
            return a/b
         except ZeroDivisionError:
            print("Non puoi dividere per 0")

a, b, op = None, None, None

while(a==None):
    try:
        a = int(input("Inserisci il primo numero: "))
    except ValueError:
        print("Numero non valido")

while(b==None):
    try:
        b = int(input("Inserisci il secondo numero: "))
    except ValueError:
        print("Numero non valido")

while(op!=None):
    op = input("Inserisci l'operatore: ")
    assert(op in ["-","+","*","/"]), "Operatore non valido"

print(calculator(a,b,op))
```
